```python
import warnings
warnings.filterwarnings('ignore')   # to avoid warnings

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

```python
tips = pd.read_csv('tips.csv')
```

```python
tips.head()
```

```python
tips.shape
```

```python
tips.info()
```

```python
# understanding the distribution of tip using distplot and boxplot
sns.distplot(tips['tip'])
plt.show()
# sns.boxplot(tips['tip'],orient='v') # Note that orient = 'v' changes the axel of the boxplot
sns.boxplot(tips['tip'])
plt.show
```

```python
sns.distplot(tips['total_bill'])
plt.show()
sns.boxplot(tips['total_bill'])
plt.show()
```

```python
# We can see above that total bill is skilled to the right with 50 percentile paying > ~ 18
```

```python
# Relationship between Total_bill and Tip
sns.jointplot(tips['total_bill'], tips['tip'],color='green')
```

```python
# We can see above that there is a strong relationship between the total bill payed and tip. 
# The Higher the bill, the higher the tip
```

```python
# Relationship using Hex Joinplot
sns.jointplot(tips['total_bill'], tips['tip'], kind="hex", color='blue')
```

```python
sns.pairplot(tips[['total_bill','tip']]);
```

```python
# Draw Linear Regression based on Tip and Total_bill
sns.lmplot(y="tip", x="total_bill", data=tips);
```

```python
# Distribution of tip per gender using stripplot

sns.stripplot(tips['sex'],tips['tip'])
```

```python
# Distribution of tip per gender using boxplot

sns.boxplot(tips['sex'],tips['tip'])
```

```python
# Linear Regression with plots by gender. Relationship between tip & total_bill per Gender

sns.lmplot(y="tip", x="total_bill",hue="sex", data=tips);
```

```python
# What day brings higher bill amount
sns.stripplot(tips['day'], tips['total_bill'], jitter=True);
```

```python
# show distribution of data point
# sns.swarmplot(tips['day'],tips['total_bill']); OR clearly defining X and Y
sns.swarmplot(x='day', y='total_bill', data=tips);
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
# Countplot 
sns.countplot(tips['day'],hue=tips['sex'])
plt.show()
```

```python
# At what time of which day does the restaurant get higher amount of bill
sns.lmplot(x='total_bill',y='tip',data=tips, fit_reg=False,hue='day', row='sex')
plt.show()
```

```python
# Plot graphs for smoker and non-smoker as it relates to tip
sns.stripplot(tips['smoker'], tips['tip'], jitter=True);
```

```python
sns.boxplot(tips['smoker'], tips['tip'])
```

```python
# Median tip of smoker is almost the same as that of non-smokers. 
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,fit_reg=False,hue='sex',col='smoker',palette='rocket')
plt.show()
```
