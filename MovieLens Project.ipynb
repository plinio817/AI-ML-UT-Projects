### 8. Visualize how popularity of genres has changed over the years - 10 marks

Note that you need to use the **percent of number of releases in a year** as a parameter of popularity of a genre

Hint 1: You need to reach to a data frame where the release year is the index and the genre is the column names (one cell shows the number of release in a year in one genre) or vice versa. (Drop unnecessary column if there are any)

Hint 2: Find the total number of movies release in a year(use item dataset to get count of movies released in a particular year, store that value in a new column as 'total'). Now divide the value of each genre in that year by total to get percentage number of release in a particular year.
`(df.div(df['total'], axis= 0) * 100)`

Once that is achieved, you can either use univariate plots or can use the heatmap to visualise all the changes over the years 
in one go. 

Hint 3: Use groupby on the relevant column and use sum() on the same to find out the number of releases in a year/genre.

```python
# Let's create a new dataframe
df_Movies2 = df_Movies.copy()
df_Movies2[['Day','Month','Year']] = df_Movies2['release date'].str.split('-',expand=True)

df_Movies2 = df_Movies2.drop(['movie title','movie id','release date','Day','Month'], axis=1)
df_Movies2.set_index('Year')
df_Movies2

# Create a new Dataframe grouping the Genre by Year
df_Movie5 = df_Movies2.groupby(by=['Year'])['Action','Adventure','Animation','Childrens','Comedy','Crime',
                                           'Documentary','Drama','Fantasy','Film-Noir','Horror',
                                           'Musical','Mystery','Romance','Sci-Fi','Thriller','War','Western'].sum()

# Create a new column to depict the sum of Genres for a given year
df_Movie5 ['Total Year'] = df_Movie5.sum(axis=1)

# Create a new dataframe to account for the % of each Gender for a given year
df_Movie6 = (df_Movie5.div(df_Movie5['Total Year'], axis= 0) * 100)

print('******* Years by Genre % ***********')
print(df_Movie6)

# Let's create the heatmap

plt.figure(figsize=(20,15))
sns.heatmap(df_Movie6, annot=True, linewidths=.5, fmt= '.1f', center = 1 )  # heatmap
plt.title('Genre Heatmap')
plt.show()


```

**Insights:**
<h4> We can see on the table and heatmap that Drama Genre is by far the most popular Genre in almost every year. </h4>
<h4> we can note in particularly, that between 1984 and 1988 (most recent years), Drama was the most popular Genre </h4>


### 9. Find the top 25 movies in terms of average ratings for movies that have been rated more than 100 times - 10 marks

Hints : 

1. Find the count of ratings and average ratings for every movie.
2. Slice the movies which have ratings more than 100.
3. Sort values according to average rating such that movie which highest rating is on top.
4. Select top 25 movies.
5. You will have to use the .merge() function to get the movie titles.

Note: This question will need you to research about groupby and apply your findings. You can find more on groupby on https://realpython.com/pandas-groupby/.

```python
# Let's create a working dataframe from the Movies to contain movie id and movie title
df_Movie2 = df_Movies[['movie id', 'movie title']]

# Let's merge Ratings with Movie title
df_Merged = pd.merge(df_Ratings,df_Movie2,how='outer',on='movie id')
df_Merged

# df_Agg = df_Merged.groupby(['movie title'].agg({'rating': ['count','mean']})
# df_Agg.to_csv('Agg.csv')

df_Agg = df_Merged.groupby(['movie title'])
counts = df_Agg.size().to_frame(name='counts')
df_Agg2 = (counts
     .join(df_Agg.agg({'rating': 'count'}).rename(columns = {'rating': 'rating_count'}))
     .join(df_Agg.agg({'rating': 'mean'}).rename(columns = {'rating': 'rating_mean'}))
     .reset_index()
)

# Create a dataframe with movies that contain > 100 ratings
df_Agg3 = df_Agg2[df_Agg2['counts'] > 100]

# Drop Count column since this is a redundant column

df_Agg3.drop(['counts'], axis='columns',inplace=True)

print('')
print('******* Top 25 Movies by Rating Count > 100 and Mean in the top 25 by Movie Tile *****')
print('')

# 25 Largest rating_mean
df_Agg3.nlargest(25,['rating_mean'])

```

### 10. Check for the validity of the below statements with respect to the data provided - 10 marks

* Men watch more drama than women
* Women watch more Sci-Fi than men
* Men watch more Romance than women

**compare the percentages**

**Please pay attention to what should be the denominator while calculating percentages**

1. Merge all the datasets

2. There is no need to conduct statistical tests around this. Just **compare the percentages** and comment on the validity of the above statements.

3. you might want ot use the .sum(), .div() function here.

4. Use number of ratings to validate the numbers. For example, if out of 4000 ratings received by women, 3000 are for drama, we will assume that 75% of the women watch drama.

```python

# Let's merge Ratings with Movie title
df_Merged1 = pd.merge(df_Ratings,df_Movies,how='outer',on='movie id')
df_Merged1

# Now Let's merge the merged Ratings and Movie with the Users file
df_Merged2 = pd.merge(df_Merged1,df_Users,how='outer',on='user id')
df_Merged2.head(3)
df_Merged2.to_csv('Merged2.csv')

# Let's create a new dataframe grouping by gender
df_Merged3 = df_Merged2.groupby(by=['gender'])['Action','Adventure','Animation','Childrens','Comedy','Crime',
                                           'Documentary','Drama','Fantasy','Film-Noir','Horror',
                                           'Musical','Mystery','Romance','Sci-Fi','Thriller','War','Western'].sum()
# print(df_Merged3)

# Create a new column in the df_Merged3 datafram to account for the sum of gender/genre per movie
df_Merged3 ['Total'] = df_Merged3.sum(axis=1)
df_Merged3.to_csv('Merged3.csv')
# print(df_Merged3)

df_Merged4 = (df_Merged3.div(df_Merged3['Total'], axis= 0) * 100)
print(df_Merged4)
df_Merged4.to_csv('Merged4.csv')

# Let's plot the data in order to answer the questions

# First let's add gender as an index of the df_Merged4 dataframe
df_Merged4['gender'] = df_Merged4.index

# Let's plot what Gender watched more Drama
print('')
plt.title ('Did Men watched more Drama than Women? as % of Total')
# plt.figure(figsize=(8,4))
ax = sns.barplot(data=df_Merged4, x="gender", y="Drama")
plt.show()

# Did Women watch more Sci-Fi than men?
print('')
plt.title ('Did Women watch more Sci-Fi as % of Total')
ax = sns.barplot(data=df_Merged4, x="gender", y="Sci-Fi")
plt.show()

# Did Men watch more Romance than Women?
print('')
plt.title('Did Men whatch more Romance as % of Total')
ax = sns.barplot(data=df_Merged4, x="gender", y="Romance")
plt.show()

```
